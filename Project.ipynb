```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

#### GETTING THE REQUIRED DATA FROM 'data.csv' file


```python
dataset = pd.read_csv('data.csv')
```

#### REPLACING ALL NULL VALUES WITH THE MEAN OF EACH COLUMN AND REMOVING ALL DUPLICATES

With the fillna() method, we may replace null values with the corresponding column's mean, median, or mode. This is a popular strategy and is helpful when the dataset only contains a few null values, as in the case of this dataset, where we will utilize the mean value and the drop duplicate() method to eliminate all duplicates.

```python
dataset = dataset.fillna(dataset_mean).drop_duplicates()
```

```python
dataset
```

```python
dataset.info()
```

```python
# Now there are no null values
dataset.isnull().sum()
```

#### THE MEAN FOR EACH COLUMN


```python
meanFA = dataset.fixed_acidity.mean()
meanVA = dataset.volatile_acidity.mean()
meanCA = dataset.citric_acid.mean()
meanRS = dataset.residual_sugar.mean()
meanC = dataset.chlorides.mean()
meanFSD = dataset.free_sulfur_dioxide.mean()
meanTSD = dataset.total_sulfur_dioxide.mean()
meanD = dataset.density.mean()
meanPH = dataset.pH.mean()
meanS = dataset.sulphates.mean()
meanA = dataset.alcohol.mean()
meanQ = dataset.quality.mean()
indexing = dataset.index
```

```python
print(f'The mean value of fixed_acidity  is {meanFA}')
print(f'The mean value of volatile_acidity is {meanVA}')
print(f'The mean value of citric_acid is {meanCA}')
print(f'The mean value of residual_sugar is {meanRS}')
print(f'The mean value of chlorides is {meanC}')
print(f'The mean value of free_sulfur_dioxide is {meanFSD}')
print(f'The mean value of total_sulfur_dioxide is {meanTSD}')
print(f'The mean value of density is {meanD}')
print(f'The mean value of pH is {meanPH}')
print(f'The mean value of sulphate is {meanS}')
print(f'The mean value of alcohol is {meanA}')
print(f'The mean value of quality is {meanQ}')
```

## GRAPHING THE DATA


Through the graphs, we can visually observe certain hidden assumptions. The dataset's variable distributions, outlines, and relationships are displayed using bar graphs, pie charts, and histograms.

```python

colors = ['indigo', 'orange', 'yellow', 'green', 'blue', 'purple',
          'pink', 'tan', 'gray', 'black', 'violet', 'beige']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5, 0, 0.5)
mean_values = [meanFA, meanVA, meanCA, meanRS, meanC,
              meanFSD, meanTSD, meanD, meanPH, meanS, meanA, meanQ]
mean_topics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
x_title = 'Column names'
y_title = 'Mean Values'
legendLabels = ['fixed_acidity (FA)',
                'volatile_acidity (VA)',
                'citric_acid (CA)',
                'residual_sugar (RS)',
                'chlorides (C)',
                'free_sulfur_dioxide (FSD)',
                'total_sulfur_dioxide (TSD)',
                'density',
                'pH',
                'sulphates',
                'alcohol',
                'quality']

labels = mean_topics
sizes = mean_values
```

```python
# plotting the pie chart
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colors, labeldistance=1.2)
plt.axis('equal')
plt.legend(title='index', labels=legendLabels, loc='upper left',)
plt.title('Mean Values for each of the columns in a pie chart')
plt.figure(figsize=(15, 18))
plt.show()
```

```python
# plotting a bar graph

fig = plt.figure(figsize=(10, 5))

plt.bar(labels, sizes, width=0.5, color=colors)
plt.xlabel(x_title)
plt.ylabel(y_title)
plt.title('Comparing mean of columns in a bar chart')
plt.legend(loc='upper right', )
plt.show()
```

#### DATA CORRELATION

We are given the correlation coefficient between each set of column pairs in the dataset by the correlation matrix. There is a positive correlation between the variables if the correlation coefficient is positive; if it is negative, there is a negative correlation between the variables. When the coefficient is 0, no correlation is observed.
